```python
train = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
test = pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
train_tr, train_val = train_test_split(train, test_size=0.2, random_state=42)
```

```python
train_tr
```

```python
knn = KNeighborsClassifier()
svm_clf = SVC(kernel="linear")
```

```python
X_train_tr = train_tr.drop(columns=['label'], axis=1)
y_train_tr = train_tr['label']
```

```python
knn.fit(X_train_tr, y_train_tr)
svm_clf.fit(X_train_tr, y_train_tr)
```

```python
X_train_val = train_tr.drop(columns=['label'], axis=1)
y_train_val = train_tr['label']
```

```python
pred1 = knn.predict(X_train_val)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_train_val, pred1)
```

```python
pred2 = knn.predict(test)
```

```python
finalSubmission = pd.DataFrame({"ImageId":0, "Label":pred2})
```

```python
finalSubmission['ImageId'] = finalSubmission.index + 1
```

```python
finalSubmission.to_csv('/kaggle/working/output.csv',index=False)
```
